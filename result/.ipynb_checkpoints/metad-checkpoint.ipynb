```python
df = pd.read_csv("/Users/zhengyuanrui/CNN_MetaCog/data/sub/sub1_metad.csv")
df_cnn = pd.read_csv("/Users/zhengyuanrui/CNN_MetaCog/data/sub/metad_cnn.csv")
```

```python
from metadpy.utils import trials2counts
nR_S1, nR_S2 = trials2counts(data=df, stimuli="stim", responses="response",accuracy="acc", confidence="rating", nRatings=2)
CNN_nR_S1, CNN_nR_S2 = trials2counts(data=df_cnn, stimuli="stim", responses="response",accuracy="acc", confidence="rating", nRatings=2)
```

```python
df_cnn
```

```python
metadpy.mle.fit_metad(nR_S1=nR_S1, 
                      nR_S2=nR_S2, 
                      nRatings=2
                      )
```

```python
metadpy.mle.fit_metad(nR_S1=CNN_nR_S1, 
                      nR_S2=CNN_nR_S2, 
                      nRatings=2
                      )
```
